```python
import pandas as pd
import numpy as np

file_path = r'C:\Users\user\OneDrive\문서\VSSource\VSPython\0_continuous.csv'
```

```python
raw_data = pd.read_csv(file_path,
                   encoding='euc-kr').values.tolist()
data = []

for elem in raw_data:
    data.append(elem[0].split('\t'))

frqunecy = []
for elem in data:
    frqunecy.append(elem[1])
```

```python
n = 108603

x = np.linspace(0, 16, n)
y = np.array(frqunecy[:108603]).astype('float64')

print(len(x))
print(len(y))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(x[:100], y[:100])
plt.show()
```
